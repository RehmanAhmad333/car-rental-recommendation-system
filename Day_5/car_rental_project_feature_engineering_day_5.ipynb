# Step 1 - Load clean dataset 

```python
import pandas as pd
```

```python
df = pd.read_csv("car_dataset.csv")
```

```python
df
```

# Step 2 - Category Encoding


```python
from sklearn.preprocessing import OneHotEncoder
```

```python
encoder = OneHotEncoder()
```

```python
category_encoding = encoder.fit_transform(df[["category"]]).toarray()
```

```python
print(category_encoding)
```

# Step 3 - Price Normalization

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scale = MinMaxScaler()
```

```python
price_scale = scale.fit_transform(df[["pricePerDay"]])
```

```python
print(price_scale)
```

# Step 4 - Feature Matrix

```python
import numpy as np
```

```python
feature_matrix = np.hstack([category_encoding , price_scale])
```

```python
print(feature_matrix.shape)
```

```python
print(feature_matrix)
```

# Save all 

```python
feature_df = pd.DataFrame(feature_matrix , columns=['2-Seater', 'Electric','Luxury', 'SUV', 'price_scaled'])
```

```python
feature_df['car_id'] = df['car_id'].values
feature_df['name'] = df['name'].values
```

```python
feature_df.to_csv("feature_matrix.csv", index=False)
```
