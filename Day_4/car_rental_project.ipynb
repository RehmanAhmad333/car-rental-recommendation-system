```python
sns.countplot(x='category', data=cr_df)
plt.title("Cars per Category")
plt.show()
```

```python
brand_avg = cr_df.groupby('brand')['pricePerDay'].mean().sort_values(ascending=False)
sns.barplot(x=brand_avg.values, y=brand_avg.index)
plt.title("Average Price by Brand")
plt.xlabel("Average Price Per Day ($)")
plt.show()
```

```python
brand_category = cr_df.groupby(['brand', 'category']).size()
print(brand_category)
```

```python
brand_cat = cr_df.groupby(['brand', 'category']).size().unstack(fill_value=0)
brand_cat.plot(kind='bar', stacked=True, figsize=(14, 6))
plt.title("Brand vs Category Distribution")
plt.xlabel("Brand")
plt.ylabel("Number of Cars")
plt.legend(title="Category")
plt.xticks(rotation=45)
plt.show()
```

```python
def assign_bucket(price):
    if price < 100:
        return "Budget"
    elif price <= 300:
        return "Mid-Range"
    else:
        return "Premium"
```

```python
cr_df['price_bucket'] = cr_df['pricePerDay'].apply(assign_bucket)
```

```python
print(cr_df[['name', 'pricePerDay', 'price_bucket']].tail(15))
```

```python
print(cr_df['price_bucket'].value_counts())
```

```python
sns.countplot(x='price_bucket', data=cr_df, 
              order=['Budget', 'Mid-Range', 'Premium'])
plt.title("Cars by Price Bucket")
plt.xlabel("Price Bucket")
plt.ylabel("Number of Cars")
plt.show()
```

```python
bucket_cat = cr_df.groupby(['price_bucket', 'category']).size().unstack(fill_value=0)
print(bucket_cat)
```
